## 14. YOLO 객체 탐지

```python
import cv2
from ultralytics import YOLO
import numpy as np
```

```python
model = YOLO("yolo11n.pt")

img = cv2.imread("../images/parrot.jpg")

results = model.predict(img, conf=0.5) # 신뢰도 0.5이상의 감지된 객체 표시
annotated_frame = results[0].plot()

cv2.imshow("img", annotated_frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#실습4 캠화면에 적용
model = YOLO("yolo11n.pt")

cap = cv2.VideoCapture(0)

while True :
    ret, frame = cap.read()

    if not ret :
        break
    
    results = model.predict(frame, conf=0.5) # 신뢰도 0.5이상의 감지된 객체 표시
    annotated_frame = results[0].plot()

    cv2.imshow("camera", annotated_frame)
    if cv2.waitKey(1) == ord("q") : 
        break
cv2.destroyAllWindows()
```
