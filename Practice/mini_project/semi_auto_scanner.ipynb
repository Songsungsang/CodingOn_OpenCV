```python
# 첫 점을 찍은 이후 마우스를 따라다니는 점을 생성하는 코드

# 스캔에 사용할 이미지를 입력합니다
img = cv2.imread("../images/card.jpg")
# 포인트를 저장할 리스트를 생성합니다
point_list = []

# 최종 결과를 출력하는 함수입니다
def show_result():
  width, height = 600, 400
  src = np.array(point_list, dtype=np.float32)
  dst = np.array([[0,0], [width, 0], [width, height], [0, height]], dtype=np.float32)
  mat = cv2.getPerspectiveTransform(src, dst)
  result = cv2.warpPerspective(img, mat, (width, height))
  cv2.imshow("result", result)

# 마우스 이벤트와 함께 실행되는 콜백함수입니다.
drawing = False
def onMouse(event, x, y, flags, _):
  copied = img.copy()
  global drawing
  if event == cv2.EVENT_LBUTTONDOWN:
    drawing = True
    #포인트 리스트에 추가하고
    point_list.append((x,y))
  
  #점을 찍을 시점부터 그리기 시작
  if drawing:
    prev_point = None
    for point in point_list: # For 문을 통해 이미지마다 점과 선 그리기 반복
      cv2.circle(copied, point, 5, (0, 0, 255), cv2.FILLED)
      if prev_point: # 처음에 None 이므로 첫 점은 선 없음
        cv2.line(copied, prev_point, point, (0,0,255), 5)
      prev_point = point # 이전 점을 현재 점으로 찍음
      
    #점을 찍고 다음 점을 찍을때까지 
    next_point = (x,y)
    cv2.line(copied, prev_point, next_point, (0, 0, 255), 5) # 마우스를 따라가는 선을 생성


  # 조건을 만족하면 결과를 출력하는 함수를 실행합니다.
  if len(point_list) == 4:
    next_point = point_list[0]
    cv2.line(copied, point_list[-1], next_point, (0, 0, 255), 5) # 마지막 선은 마지막 점과 처음 점을 이어줌
    show_result()
  
  cv2.imshow(name, copied) # 마우스 움직임마다 사진을 새로 출력

# 기본적인 코드 실행입니다.
name = "Scanner"
cv2.namedWindow(name)
cv2.setMouseCallback(name, onMouse)
cv2.imshow(name, img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
