## 02. 비디오 입출력

### 2-1. 비디오 파일 입출력
- 비디오의 재생 : 연속된 이미지를 재생하는 것
- 프레임 : 연속된 이미지 중 하나
- 비디오를 재생하기 위해서는 프레임을 연속적으로 재생 = 사진을 연속으로 띄워주기

```python
import cv2
PARROT_PATH = "videos/parrot.mp4"
KELSY_PATH = "videos/kelsy.gif"
```

### 2-1. 비디오 파일 입출력
- cv2.VideoCapture("파일경로") : 경로의 파일을 불러옴
- VideoCapture.isOpened() : 파일이 열렸는지 확인
- VideoCapture.read() : 파일을 읽어서 frame을 반환
    - ret : 프레임이 있는지 체크, frame : 불러온 이미지
- VideoCapture.release() : 사용한 영상자원 해제

```python
cap = cap = cv2.VideoCapture(KELSY_PATH)
print(cap)
```

```python
cap = cv2.VideoCapture(PARROT_PATH)
print(cap)
```

```python
cap = cv2.VideoCapture(PARROT_PATH)
print(cap.isOpened())
```

```python
cap = cv2.VideoCapture(PARROT_PATH)

while cap.isOpened(): # 영상이 열려있다면
    ret, frame = cap.read()
    print(ret, frame)
    
    if not ret: # 리턴값이 없어
        break # 그럼 탈출

    cv2.imshow("video", frame) # 프레임을 계속 보여줌 = 영상

    if cv2.waitKey(20) == ord("q") : # q를 입력하면 탈출. 또한 프레임과 프레임 사이의 재생 속도를 결정하게됨
        break

cap.release() # 사용한 영상자원 해제
cv2.destroyAllWindows()
```

### 2-2. 영상의 프레임 불러오기
- VideoCapture.get(cv2.CAP_PROP_FPS)

```python
cap = cv2.VideoCapture(PARROT_PATH)
fps = cap.get(cv2.CAP_PROP_FPS)
print(fps)
```

```python
cap = cv2.VideoCapture(PARROT_PATH)
fps = cap.get(cv2.CAP_PROP_FPS)
print(fps)

while cap.isOpened() :
    ret, frame = cap.read()

    if not ret:
        break

    cv2.imshow("video", frame)

    if cv2.waitKey(int(1000/fps)) == ord("q") :
        break

cap.release() # 절대로 반복문 안에 넣지 말도록
cv2.destroyAllWindows()
```

```python
# 웹캠 띄우기
cap = cv2.VideoCapture(0)

if not cap.isOpened() :
    exit()

while True :
    ret, frame = cap.read()

    if not ret :
        break

    cv2.imshow("Camera", frame)

    if cv2.waitKey(1) == ord("q") :
        break

cap.release()
cv2.destroyAllWindows()
```

## 2-4 비디오 쓰기
 - cv2.VideoWriter(파일경로, fourcc, fps, (가로, 세로)) : 영상을 쓰기

```python
# 영상을 프레임마다 읽어서 이미지 캡쳐하고 그걸 합쳐서 다른 파일로 저장

cap = cv2.VideoCapture(PARROT_PATH)
fourcc = cv2.VideoWriter_fourcc(*"DIVX")
fps = cap.get(cv2.CAP_PROP_FPS)
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # 가로와 세로가 정수값을 가져야 하므로 반올림 함
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter("output/parrot.mp4", fourcc, fps, (width, height)) # 비디오 쓰기를 위해 필요 정보 입력. fourcc - 영상 인코딩 방식

while cap.isOpened() :
    ret, frame = cap.read()

    if not ret : 
        break
    
    out.write(frame)
    cv2.imshow("video", frame)

    if cv2.waitKey(int(1000/fps)) == ord("q") : 
        break

out.release()
cap.release()
cv2.destroyAllWindows()
```

```python
# 실습 - 웹캠 화면을 특정 키 입력시 찍고 카메라를 끄도록

# 웹캠 띄우기
cap = cv2.VideoCapture(0)

if not cap.isOpened() :
    exit()

while True :
    ret, frame = cap.read()

    if not ret :
        break

    cv2.imshow("Camera", frame)

    if cv2.waitKey(1) != -1 : # -1 = 입력값이 없음. 즉, 어떤 입력값이라도 받았다면
        cv2.imwrite("output/webcam_capture.png", frame) # 키를 누르면 그 프레임 이미지를 write
        break

cap.release()
cv2.destroyAllWindows()
```
